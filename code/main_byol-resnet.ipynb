```python
from MRIBTCPB.code.train import train_model, train_byol, train_classifier
from MRIBTCPB.code.optuna_train import run_experiments
from MRIBTCPB.code.utils.utils import plot_tsne_representation
```

```python
model = 'resnet50'
batch_size = 128

Byol = train_byol(model, batch_size, epochs=30)
```

```python
byol_resnet50_params = run_experiments(Byol.model, fine_tune=True, byol=True)
```

```python
classifier = train_classifier(Byol, fine_tune=True, **byol_resnet50_params)
```

```python
plot_tsne_representation(classifier, n_components=2)
```
